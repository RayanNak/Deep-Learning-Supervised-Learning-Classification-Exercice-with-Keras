```python
# Importing modules which we're gonna need

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing the dataset and visualize it

from sklearn.datasets import make_circles

X, y = make_circles(n_samples=500, noise=0.01, random_state=500000)
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.legend(['0', '1'])
plt.show()
```

```python
# Importing the Deep Learning Framework we're gonna use which is Keras

import keras
from keras.layers import Dense
from keras.optimizers import Adam
from keras.models import Sequential
```

```python
# Creating a very simple neural network for the beginning

model = Sequential()
model.add(Dense(1, input_shape=(2,), activation='sigmoid'))
model.compile(Adam(lr=0.01), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=100, verbose=1)
```

```python
# Creating the 'plot_decision_boundary' function for visualization

def plot_decision_boundary(model, X, y):
    amin, bmin = X.min(axis=0) - 0.1
    amax, bmax = X.max(axis=0) + 0.1
    hticks = np.linspace(amin, amax, 101)
    vticks = np.linspace(bmin, bmax, 101)
    
    aa, bb = np.meshgrid(hticks, vticks)
    ab = np.c_[aa.ravel(), bb.ravel()]
               
    c = model.predict(ab)
    cc = c.reshape(aa.shape)
    
    plt.figure(figsize=(12, 8))
    plt.contourf(aa, bb, cc, cmap='bwr', alpha=0.2)
    plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
    plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
    plt.legend(['1', '0'])

plot_decision_boundary(model, X, y)
```

```python
# Creating a Deeper neural network than the first one

model = Sequential()
model.add(Dense(6, input_shape=(2,), activation='tanh'))
model.add(Dense(6, activation='tanh'))
model.add(Dense(6, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.compile(Adam(lr=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#  Fitting our model

h = model.fit(X, y, epochs=250, verbose=1)
```

```python
# Visualization

plot_decision_boundary(model, X, y)
```

```python
# Storing the loss function values in a variable for visualization later

losses = h.history['loss']
epochs = [i+1 for i in range(250)]
```

```python
# Plotting the loss function

plt.plot(epochs, losses)
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```

```python
# Storing the accuracy metric in a variable for visualization later

accuracy = h.history['acc']
```

```python
# Plotting the accuracy metric

plt.plot(epochs, accuracy)
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.show()
```
